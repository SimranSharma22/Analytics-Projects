```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

# use the filename of your uploaded CSV file
df = pd.read_csv("Dataset.xlsx.csv")
```

```python
#1.How many Males/Females are there in the entire organization?
```

```python
# Count the number of occurrences of each gender
gender_counts = df['Gender'].value_counts()

# Print the results
print("Number of males:", gender_counts['Male'])
print("Number of females:", gender_counts['Female'])
```

```python
#2.How many Males/Females are there in the each department or for each location?
```

```python
# Count the number of occurrences of each gender for each department
gender_counts_by_department = df.groupby('Department')['Gender'].value_counts()

# Count the number of occurrences of each gender for each location
gender_counts_by_location = df.groupby('Loc')['Gender'].value_counts()

# Print the results
print("Gender counts by department:\n", gender_counts_by_department)
print("Gender counts by location:\n", gender_counts_by_location)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot a bar chart of the gender counts by department
plt.figure()
gender_counts_by_department.unstack().plot(kind='bar', stacked=True)
plt.title('Gender Counts by Department')
plt.xlabel('Department')
plt.ylabel('Count')

# Plot a bar chart of the gender counts by location
plt.figure()
gender_counts_by_location.unstack().plot(kind='bar', stacked=True)
plt.title('Gender Counts by Loc')
plt.xlabel('Loc')
plt.ylabel('Count')

plt.show()
```

```python
#3 For which department is the average Pay highest?
```

```python
# Remove dollar signs and commas from the 'Salary' column
df['Salary'] = df['Salary'].str.replace('$', '').str.replace(',', '')

# Convert the 'Salary' column to a float df type
df['Salary'] = df['Salary'].astype(float)

# Calculate the mean salary for each department
avg_salary_by_department = df.groupby('Department')['Salary'].mean()

# Find the department with the highest average salary
highest_avg_salary_department = avg_salary_by_department.idxmax()

# Print the result
print("Department with the highest average salary:", highest_avg_salary_department)
```

```python
import matplotlib.pyplot as plt

# Plot a bar chart of the average salary by department
plt.figure()
avg_salary_by_department.plot(kind='bar')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Salary')

plt.show()
```

```python
#5.What percentage of employees received good & very good rating? What about poor & very poor rating? and average rating?
```

```python
# Calculate the percentage of employees with each rating
good_rating_percentage = len(df[df['Rating'] == 'Good']) / len(df) * 100
very_good_rating_percentage = len(df[df['Rating'] == 'Very Good']) / len(df) * 100
poor_rating_percentage = len(df[df['Rating'] == 'Poor']) / len(df) * 100
very_poor_rating_percentage = len(df[df['Rating'] == 'Very Poor']) / len(df) * 100
average_rating_percentage = len(df[df['Rating'] == 'Average']) / len(df) * 100

# Print the results
print("Percentage of employees with Good rating:", good_rating_percentage)
print("Percentage of employees with Very Good rating:", very_good_rating_percentage)
print("Percentage of employees with Poor rating:", poor_rating_percentage)
print("Percentage of employees with Very Poor rating:", very_poor_rating_percentage)
print("Percentage of employees with Average rating:", average_rating_percentage)
```

```python
# Plot a pie chart of the rating percentages
plt.figure()
labels = ['Good', 'Very Good', 'Poor', 'Very Poor', 'Average']
sizes = [good_rating_percentage, very_good_rating_percentage, poor_rating_percentage, very_poor_rating_percentage, average_rating_percentage]
colors = ['green', 'limegreen', 'red', 'indianred', 'yellow']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Employee Ratings')

plt.show()
```

```python
#6.Compute gender pay gap for each department. Interpret
```

```python
# Group the data by department and gender, and calculate the mean salary for each group
avg_salary_by_department_gender = df.groupby(['Department', 'Gender'])['Salary'].mean()

# Calculate the gender pay gap for each department
gender_pay_gap_by_department = (avg_salary_by_department_gender.loc[:, 'Male'] - avg_salary_by_department_gender.loc[:, 'Female']) / avg_salary_by_department_gender.loc[:, 'Male'] * 100

# Print the results
print("Gender pay gap by department:")
print(gender_pay_gap_by_department)
```

```python
# Create a bar chart of the gender pay gap by department
plt.bar(gender_pay_gap_by_department.index, gender_pay_gap_by_department.values)
plt.xticks(rotation=90)
plt.title('Gender Pay Gap by Department')
plt.xlabel('Department')
plt.ylabel('Gender Pay Gap (%)')
plt.show()
```

```python
#Interpretation: From the above graph, we can interpret that The output shows the gender pay gap (in percentage) for different departments.
#A positive value indicates that on average, men earn more than women in that department, while a negative value indicates that on average,
#women earn more than men in that department.

#For example,
#In the department of Business Development, men earn on average 9.008776% more than women. On the other hand, in the department of
#Engineering, women earn on average 11.050498% more than men, indicating a reverse gender pay gap.

#Overall, this output can help identify departments with the largest gender pay gap
#and guide efforts to address pay equity and diversity in the workplace.
```

```python
#7. Compute gender pay gap for each location. Interpret.
```

```python
# Group the data by department and gender, and calculate the mean salary for each group
avg_salary_by_loc_gender = df.groupby(['Loc', 'Gender'])['Salary'].mean()

# Calculate the gender pay gap for each department
gender_pay_gap_by_loc = (avg_salary_by_loc_gender.loc[:, 'Male'] - avg_salary_by_loc_gender.loc[:, 'Female']) / avg_salary_by_loc_gender.loc[:, 'Male'] * 100

# Print the results
print("Gender pay gap by Location:")
print(gender_pay_gap_by_loc)
```

```python
import scipy.stats as stats

# Correlation analysis between Gender and Loc
Gender_Loc_corr = pd.crosstab(df['Gender'], df['Loc'])
chi2_Gender_Loc, p_Gender_Loc, dof_Gender_Loc, expected_Gender_Loc = stats.chi2_contingency(Gender_Loc_corr)
print('Gender-Loc Correlation: chi2=%.2f, p=%.2f' % (chi2_Gender_Loc, p_Gender_Loc))

# Correlation analysis between Gender and Department
Gender_Department_corr = pd.crosstab(df['Gender'], df['Department'])
chi2_Gender_Department, p_Gender_Department, dof_Gender_Department, expected_Gender_Department = stats.chi2_contingency(Gender_Department_corr)
print('Gender-Department Correlation: chi2=%.2f, p=%.2f' % (chi2_Gender_Department, p_Gender_Department))

# Correlation analysis between Gender and Salary
male_salaries = df.loc[df['Gender'] == 'Male', 'Salary']
female_salaries = df.loc[df['Gender'] == 'Female', 'Salary']
t_Salary, p_Salary = stats.ttest_ind(male_salaries, female_salaries)
print('Gender-Salary Correlation: t=%.2f, p=%.2f' % (t_Salary, p_Salary))

# Correlation analysis between Loc and Salary
Loc_Salary_corr = pd.crosstab(df['Loc'], pd.qcut(df['Salary'], q=3))
chi2_Loc_Salary, p_Loc_Salary, dof_Loc_Salary, expected_Loc_Salary = stats.chi2_contingency(Loc_Salary_corr)
print('Loc-Salary Correlation: chi2=%.2f, p=%.2f' % (chi2_Loc_Salary, p_Loc_Salary))

# Correlation analysis between Department and Salary
Department_Salary_corr = pd.crosstab(df['Department'], pd.qcut(df['Salary'], q=3))
chi2_Department_Salary, p_Department_Salary, dof_Department_Salary, expected_Department_Salary = stats.chi2_contingency(Department_Salary_corr)
print('Department-Salary Correlation: chi2=%.2f, p=%.2f' % (chi2_Department_Salary, p_Department_Salary))

```
